```python
import pandas as pd 
test_text_df = pd.DataFrame([u"今天天氣很好",u"今天天氣很爛",u"我恨它",u"它恨我",u"我愛它",u"它愛我",u"今天很衰",u"日子一天一天過",u"天天刷牙洗臉"])
test_text_df.columns = ["text"]
test_text_df = test_text_df.reset_index()
test_text_df["idx"] = map(lambda xx :u"%05d" % xx,test_text_df["index"])
test_text_df["r1"] = np.random.randn(test_text_df.shape[0])
test_text_df["r2"] = np.random.randn(test_text_df.shape[0])
test_text_df
```

```python
test_text_df["text"].values
```

```python
from PlaYdata.core.base import ValuesMatrix
from PlaYdata.core.base import tools
from PlaYdata.core.merge import MultiIndexedDataMatrixMerger, IndexedDataMatrixMerger
```

```python
text_val_matrix = map(lambda xx:ValuesMatrix(list(tools.ngram(xx,[1]))),test_text_df["text"].values)
```

# Bigram Example

```python
text_idxed_data_matrix = map(lambda xx:xx.build_index_data_matrix()._1d_ngram(2), text_val_matrix)
idx_merger = IndexedDataMatrixMerger(*text_idxed_data_matrix)
bigram_mat = idx_merger.rbind()
```

```python
import numpy as np
import scipy.sparse as sparse
from PlaYnlp.sparse import SparseDataFrame
from PlaYnlp.sparse import L1_norm_col_summarizer, L0_norm_col_summarizer
```

```python
col = np.arange(ii._ncol)
data = np.ones(ii._as_1d_array.shape)
row = ii._as_1d_array
```

```python
states_sdf = SparseDataFrame(smatrix=sparse.coo_matrix((data,(row,col))).T,summarizer=L1_norm_col_summarizer)
```

```python
states_sdf.summary._data
```

```python
import pandas as pd
```

```python
bigram_groupby_df = pd.DataFrame(bigram_mat.states_matrix._eval(ss),columns=("w1","w2"))
bigram_groupby_df["N"] = states_sdf.summary._data
bigram_groupby_df
```

# fourth-gram Example

```python
text_val_matrix[2].shape[1]
```

```python
text_idxed_data_matrix = [xx.build_index_data_matrix()._1d_ngram(4) for xx in text_val_matrix if xx.shape[1] >= 4]
```

```python
text_idxed_data_matrix
```

```python
idx_merger = IndexedDataMatrixMerger(*text_idxed_data_matrix)
forth_gram_mat = idx_merger.rbind()
ss, ii = forth_gram_mat.index_matrix.build_row_struct_index()
col = np.arange(ii._ncol)
data = np.ones(ii._as_1d_array.shape)
row = ii._as_1d_array
states_sdf = SparseDataFrame(smatrix=sparse.coo_matrix((data,(row,col))).T,summarizer=L1_norm_col_summarizer)
```

```python
bigram_groupby_df = pd.DataFrame(forth_gram_mat.states_matrix._eval(ss),columns=("w1","w2","w3","w4"))
bigram_groupby_df["N"] = states_sdf.summary._data
bigram_groupby_df
```
